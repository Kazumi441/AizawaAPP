```python
import sqlite3
import requests
import urllib.parse

def get_coordinates_from_address(address):
    base_url = "https://msearch.gsi.go.jp/address-search/AddressSearch"
    encoded_address = urllib.parse.quote(address)
    url = f"{base_url}?q={encoded_address}"

    response = requests.get(url)
    data = response.json()

    if data:
        coordinates = data[0]["geometry"]["coordinates"]
        latitude, longitude = coordinates[1], coordinates[0]
        return latitude, longitude
    else:
        return None, None

# データベース接続
conn = sqlite3.connect('chuko_database.db')
cursor = conn.cursor()

# データベース内の住所情報を取得
cursor.execute("SELECT id, address FROM chuko_table")
rows = cursor.fetchall()

# 住所から緯度と経度を取得してデータベースに追加
for row in rows:
    property_id, address = row
    latitude, longitude = get_coordinates_from_address(address)
    if latitude is not None and longitude is not None:
        cursor.execute("UPDATE chuko_table SET latitude=?, longitude=? WHERE id=?", (latitude, longitude, property_id))
        conn.commit()

        # 画面上表示
        print(f"住所: {address}, 緯度: {latitude}, 経度: {longitude}")
    else:
        print(f"住所: {address} の緯度と経度は取得できませんでした。")

# データベース接続を閉じる
conn.close()
```

```python
import sqlite3
import requests
import urllib.parse

def get_coordinates_from_address(address):
    base_url = "https://msearch.gsi.go.jp/address-search/AddressSearch"
    encoded_address = urllib.parse.quote(address)
    url = f"{base_url}?q={encoded_address}"

    response = requests.get(url)
    data = response.json()

    if data:
        coordinates = data[0]["geometry"]["coordinates"]
        latitude, longitude = coordinates[1], coordinates[0]
        return latitude, longitude
    else:
        return None, None

# データベース接続
conn = sqlite3.connect('chintai_database.db')
cursor = conn.cursor()

# カラムが存在しない場合は作成
cursor.execute('''
    CREATE TABLE IF NOT EXISTS my_table (
        id INTEGER PRIMARY KEY,
        所在地 TEXT,
        latitude REAL,
        longitude REAL
    )
''')
conn.commit()

# データベース内の所在地情報を取得
cursor.execute("SELECT id, 所在地 FROM my_table")
rows = cursor.fetchall()

# 緯度と経度を取得してデータベースに追加
for row in rows:
    property_id, address = row
    latitude, longitude = get_coordinates_from_address(address)
    if latitude is not None and longitude is not None:
        cursor.execute("UPDATE my_table SET latitude=?, longitude=? WHERE id=?", (latitude, longitude, property_id))
        conn.commit()

        # 画面上表示
        print(f"所在地: {address}, 緯度: {latitude}, 経度: {longitude}")
    else:
        print(f"所在地: {address} の緯度と経度は取得できませんでした。")

# データベース接続を閉じる
conn.close()
```
